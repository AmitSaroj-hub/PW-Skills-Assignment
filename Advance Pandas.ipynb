#Q2. Sales Filtering Analysis

Orders where Sales > 3500

```python
sales_above_3500 = df[df['Sales'] > 3500]
sales_above_3500
```

Orders where Region = "East" and Sales > 2500

```python
east_sales_above_2500 = df[
    (df['Region'] == 'East') &
    (df['Sales'] > 2500)
]

east_sales_above_2500
```

#Q3. Customer Purchase Analysis

```python
top_5_customers = (
    df.groupby('Customer')['Sales']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print(top_5_customers)
```

#Q4. Regional Sales Performance

```python
regional_sales = df.groupby('Region')['Sales'].sum()

print(regional_sales)

print("\nHighest Revenue Region:")
print(regional_sales.idxmax())
```

#Q5. Product Performance Analysis

```python
product_quantity = (
    df.groupby('Product')['Quantity']
      .sum()
      .sort_values(ascending=False)
)

print(product_quantity)
```

#Q6. Multi-Level Grouping

```python
region_category_sales = (
    df.groupby(['Region', 'Category'])['Sales']
      .sum()
)

print(region_category_sales)
```

#Q7. Pivot Table Creation

```python
pivot_table = pd.pivot_table(
    df,
    index='Region',
    columns='Product',
    values='Sales',
    aggfunc='sum'
)

print(pivot_table)
```

#Q8. Profitability Analysis

```python
avg_profit_category = (
    df.groupby('Category')['Profit']
      .mean()
      .sort_values(ascending=False)
)

print(avg_profit_category)
```

#Q9. Discount Impact Study

```python
discount_profit = (
    df.groupby('Discount')['Profit']
      .mean()
)

print(discount_profit)
```

#Q10. End-to-End Data Wrangling Task

```python
filtered_df = df[df['Sales'] > 3000]

grouped_sales = (
    filtered_df.groupby(['Region', 'Product'])['Sales']
    .sum()
    .reset_index()
)

pivot_result = grouped_sales.pivot(
    index='Region',
    columns='Product',
    values='Sales'
)

print(pivot_result)
```
